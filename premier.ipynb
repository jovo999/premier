```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Premier.csv')
df.head()
```

```python
df.drop('*Additional Stats', axis=1, inplace=True)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Attendance'] = df['Attendance'].str.replace(',' , '')
df['Attendance'] = df['Attendance'].str.replace('NaN' , '0')
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
df['Attendance'] = df['Attendance'].astype(float)
```

```python
df.isnull().sum()
```

```python
cols = ['homeScore' , 'homeXG' , 'awayScore' , 'awayXG' , 'Attendance']
df.loc[:, cols] = df.loc[:, cols].fillna(df.loc[:, cols].mean())
df.head()
```

```python
df['Time'].value_counts().head()
```

```python
df['Referee'] = df['Referee'].fillna('uknowe')
df['Result'] = df['Result'].fillna('uknowe')
```

```python
df.isnull().sum()
```

```python
HTS = df.groupby('Home Team')['homeScore'].sum().reset_index().sort_values(by='homeScore', ascending=False ).round(2)
HTS.head(10)
```

```python
HTH = df.groupby('Home Team')['homeXG'].sum().reset_index().sort_values(by='homeXG', ascending=False ).round(2)
HTH.head(10)
```

```python
df['Stadium'].value_counts().head()
```

```python
sum_of_Attendance = df.groupby('Stadium')['Attendance'].sum().reset_index().sort_values(by='Attendance' , ascending=False).round(2)
sum_of_Attendance.head()
```

```python
df['Result'].value_counts()
```

```python
reef = df['Referee'].value_counts().reset_index().head(10)
reef
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Home Team', y='homeScore', data=HTS.head(10) , palette='vlag')
plt.title('Home Team Scores')
plt.xlabel('Home Team')
plt.ylabel('Score')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Home Team', y='homeXG', data=HTH.head(10) , palette='vlag')
plt.title('Home Team Scores')
plt.xlabel('Home Team')
plt.ylabel('Score')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = reef , x= 'Referee' , y= 'count' , palette='vlag')
plt.title('Referee')
plt.xlabel('Referee')
plt.xticks(rotation=90)
plt.show()
```

```python
num = df.select_dtypes(include='number')

plt.figure(figsize=(10, 6))
sns.heatmap(num.corr() , annot=True , cmap='vlag')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = sum_of_Attendance , x= 'Stadium' , y= 'Attendance' , palette='vlag')
plt.title('sum of Attendance')
plt.xlabel('Stadium')
plt.xticks(rotation=90)
plt.show()
```
